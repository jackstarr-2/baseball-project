### Test whether teams that made the playoffs hit significantly more home runs than teams that missed

```python
from scipy.stats import ttest_ind

# Split HomeRuns into playoff and non-playoff groups
playoff_hrs = df[df['MadePlayoffs'] == 'Y']['HomeRuns']
nonplayoff_hrs = df[df['MadePlayoffs'] == 'N']['HomeRuns']

# Perform independent samples t-test
t_stat, p_val = ttest_ind(playoff_hrs, nonplayoff_hrs, equal_var=False)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_val}")
```

#### I can conclude that playoff teams hit significantly more homeruns on average than non playoff teams

### Predict Playoff qualification from Home Runs

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# Predictor and target
X = df[['HomeRuns']]
y = df['MadePlayoffs'].map({'Y': 1, 'N': 0})

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train model
model = LogisticRegression()
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# Evaluate
print(classification_report(y_test, y_pred))
```

#### This suggests that while homeruns are associated with postseason success, home runs alone are not enough to reliably predict which teams will make the playoffs

### Predicting wins from homeruns, runs, and hits

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Predictor variables and target variable
X = df[['HomeRuns', 'Runs', 'Hits', 'ERA','StrikeoutsPitching', 'HRAllowed', 'HR_per_Game', 'HR_diff']]
y = df['Wins']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
print("R-squared:", r2_score(y_test, y_pred))
print("Root Mean Squared Error:", mean_squared_error(y_test, y_pred, squared=False))
```

#### A simple linear regression showed that about 87% of team wins could be explained by offensive stats and defensive stats as well as HR_diff and HR_per_Game. It has a mean squared error of 4.32 which is pretty accurrate

## Question 3: Predicting the amount of home runs each team will hit in the 2025 season

```python
from sklearn.ensemble import RandomForestRegressor

# Define your selected features
features = ['HRAllowed', 'Runs', 'ERA', 'HR_diff', 'HR_per_Game', 'Losses', 'Wins']

# Use all data before 2024 to train
train_df = df[df['Year'] < 2024]
X_train = train_df[features]
y_train = train_df['HomeRuns']

# Use 2024 stats to predict 2025
X_2024 = df[df['Year'] == 2024][features]
teams_2024 = df[df['Year'] == 2024][['TeamID']].reset_index(drop=True)

# Train model
model = RandomForestRegressor(random_state=42)
model.fit(X_train, y_train)

# Predict 2025 HRs
predicted_hr_2025 = model.predict(X_2024)
predicted_df = teams_2024.copy()
predicted_df['Predicted_HR_2025'] = predicted_hr_2025.round(1)

# Output predictions
print(predicted_df)
```

```python
# Sort the predictions from most to least home runs
predicted_df = predicted_df.sort_values(by='Predicted_HR_2025', ascending=False)

# Optional: reset index if you want a clean print
predicted_df = predicted_df.reset_index(drop=True)

# Now you can print or plot it
print(predicted_df)
```

```python
# Step 1: Merge in Conference and Division info first
team_info = df[df['Year'] == 2024][['TeamID', 'Conference', 'Division']].drop_duplicates()
predicted_df = predicted_df.merge(team_info, on='TeamID', how='left')

# Step 2: Rank by HRs
predicted_df['PlayoffRank'] = predicted_df['Predicted_HR_2025'].rank(method='first', ascending=False)

# Step 3: Flag top 10 as initial playoff guesses (if not using division logic yet)
predicted_df['PlayoffGuess'] = predicted_df['PlayoffRank'] <= 10

# Step 4: Sort for display
predicted_df = predicted_df.sort_values(by='Predicted_HR_2025', ascending=False).reset_index(drop=True)

# Step 5: Preview
print(predicted_df.head())
```

```python
# Start clean
df = predicted_df.copy()

def select_playoff_teams_by_league(league_df):
    # 1. Select best team from each division
    division_winners = league_df.groupby('Division').apply(
        lambda g: g.nlargest(1, 'Predicted_HR_2025')
    ).reset_index(drop=True)

    # 2. Get next 3 best (wild cards), excluding division winners
    remaining = league_df[~league_df['TeamID'].isin(division_winners['TeamID'])]
    wild_cards = remaining.nlargest(3, 'Predicted_HR_2025')  # ← changed 2 to 3

    # 3. Add a tag
    division_winners['PlayoffType'] = 'Division Winner'
    wild_cards['PlayoffType'] = 'Wild Card'

    return pd.concat([division_winners, wild_cards])

# Separate by league
al_df = df[df['Conference'] == 'AL']
nl_df = df[df['Conference'] == 'NL']

# Apply playoff selection
al_playoffs = select_playoff_teams_by_league(al_df)
nl_playoffs = select_playoff_teams_by_league(nl_df)

# Combine into final playoff dataframe
playoff_teams = pd.concat([al_playoffs, nl_playoffs]).reset_index(drop=True)

# Update original dataframe with playoff info
df['PlayoffGuess'] = df['TeamID'].isin(playoff_teams['TeamID'])
df = df.merge(playoff_teams[['TeamID', 'PlayoffType']], on='TeamID', how='left')
df['PlayoffType'] = df['PlayoffType'].fillna('Not in Playoffs')

# Optional: sort for readability
df = df.sort_values(by='Predicted_HR_2025', ascending=False).reset_index(drop=True)

# View result
print(df[['TeamID', 'Predicted_HR_2025', 'Conference', 'Division', 'PlayoffGuess', 'PlayoffType']])
```
